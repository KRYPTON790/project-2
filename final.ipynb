```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from pandas import DataFrame
```

```python
data=pd.read_csv('train.csv')
data=data.fillna(0)
data['Sex']=data.Sex.map({'female':0, 'male':1})
```

```python
x = data[["Age","Sex","Pclass"]]
y = data[["Survived"]]
```

```python
x_train =np.array(x)
y_train =np.array(y)
```

```python
test=pd.read_csv('test_X.csv')
test=test.fillna(0)
test['Sex']=test.Sex.map({'female':0, 'male':1})
```

```python
x_test = test[["Age","Sex","Pclass"]]
```

```python
result=pd.read_csv('test_Y.csv')
survived_test=pd.get_dummies(result.Survived)
```

```python
y_1= result[["Survived"]]
```

```python
y_result = np.array(y_1)
```

```python
regr = linear_model.LinearRegression()
regr.fit(x_train,y_train)
```

```python
 x_result =regr.predict(x_test)
```

```python
MSE=mean_squared_error(x_result,y_result)
print(MSE)
```
